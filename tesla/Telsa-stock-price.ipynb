Tesla stock  price is highly manipulated by inside investors, news sentiment and products launchment. I have been reseraching for a while how to make profit on its volatile behavior.

The key is to remove the data noice, smooth out the historical price and understand how much it is really worth.

```python
import pandas as pd
pd.__version__

import plotly.express as px
import warnings
warnings.simplefilter(action='ignore',category=Warning)
```

```python
import pandas_datareader as web
import datetime as dt

import matplotlib.pyplot as plt
```

```python
tesla=pd.read_csv('TSLA.csv')
tesla=tesla[::-1]
tesla.head()
```

```python
rolling_back_days=[7,14,30,60,120]

tesla['7days_rolling_back_avg']=tesla['Adj Close'].rolling(7).mean().shift(-6)
tesla['14days_rolling_back_avg']=tesla['Adj Close'].rolling(14).mean().shift(-13)
tesla['30days_rolling_back_avg']=tesla['Adj Close'].rolling(30).mean().shift(-29)
tesla['60days_rolling_back_avg']=tesla['Adj Close'].rolling(60).mean().shift(-59)
tesla['120days_rolling_back_avg']=tesla['Adj Close'].rolling(120).mean().shift(-119)
tesla['150days_rolling_back_avg']=tesla['Adj Close'].rolling(150).mean().shift(-149)
tesla['365days_rolling_back_avg']=tesla['Adj Close'].rolling(365).mean().shift(-364)
tesla['720days_rolling_back_avg']=tesla['Adj Close'].rolling(720).mean().shift(-719)
```

```python
tesla.head()
```

```python
columns=['Date','Adj Close','7days_rolling_back_avg','14days_rolling_back_avg',\
         '30days_rolling_back_avg','60days_rolling_back_avg','120days_rolling_back_avg','150days_rolling_back_avg','365days_rolling_back_avg','720days_rolling_back_avg']
```

```python
tesla_df=tesla[columns]
tesla_df['Date']=pd.to_datetime(tesla_df['Date'])
tesla_df.head(20)
```

```python
tesla_df_1=tesla_df.melt(id_vars=['Date']+list(tesla_df.keys()[2:4]),var_name=['Tesla Stock Price'])
fig=px.line(tesla_df_1,x='Date',y='value',color='Tesla Stock Price')
fig.show()
```

```python
tesla_df_1=tesla_df.melt(id_vars=['Date']+list(tesla_df.keys()[1:8]),var_name=['Tesla Stock Price'])
fig=px.line(tesla_df_1,x='Date',y='value',color='Tesla Stock Price')
fig.show()
```
